```python
from ultralytics import YOLO
import cv2
from glob import glob
import os
from pathlib import Path

model = YOLO('/home/anton/PycharmProjects/kaggle_competitions/severstal_aiijc/runs/detect/train6/weights/best.pt')

def crop_4_parts(images_path):
    crops_path = 'crops_4'
    if not Path(crops_path).exists():
        os.mkdir(crops_path)

    for img_path in images_path:
        curr_path = os.path.join(crops_path, img_path.split('/')[-1][:-4])
        if not Path(curr_path).exists():
            os.mkdir(curr_path)
        
        image = cv2.imread(img_path)
        img_size = (image.shape[0] // 2, image.shape[1] // 2)
        images = [image[:img_size[0], :img_size[1]], image[:img_size[0], img_size[1]:],
                 image[img_size[0]:, :img_size[1]], image[img_size[0]:, img_size[1]:]]
        
        for i, img in enumerate(images):
            cv2.imwrite(curr_path + '/' + str(i) + '.png', img)
        
    
images_path = glob('severstal-pipes/test/0/*') + glob('severstal-pipes/test/1/*')
crop_4_parts(images_path)

for dirr in os.listdir('crops_4'):
    path = os.path.join('crops_4', dirr)
    results = model.predict(source=path, conf=0.6, save=True)
```
